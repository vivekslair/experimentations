```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
asset_df= pd.read_csv("/users/Vivek/Documents/assets/cleansed_data.csv")
```

```python
asset_df.columns
```

```python
asset_df.drop('Unnamed: 0',inplace=True,axis=1)
```

```python
asset_df.groupby('Branch',as_index=False).agg({'Total Collections':'sum','Pretax Revenue':'sum','Gross Margin':'sum',\
                                               'Net Margin':'sum','Route Name': 'nunique','Spoiled Cost':'sum',\
                                               'Product Cost':'sum',\
                                               'Machine Inventory Cost': 'sum'}).sort_values('Total Collections',ascending=False)
```

```python
asset_df.groupby('Manufacturer',as_index=False).agg({'Total Collections':'sum','Pretax Revenue':'sum','Gross Margin':'sum',\
                                                     'Net Margin':'sum','Route Name': 'nunique','Spoiled Cost':'sum',\
                                                     'Product Cost':'sum','Machine Inventory Cost': 'sum'}).sort_values('Total Collections',ascending=False)
```

```python
asset_df.groupby(['Branch','Manufacturer'],as_index=False).agg({'Total Collections':'sum','Pretax Revenue':'sum','Gross Margin':'sum',\
                                               'Net Margin':'sum','Route Name': 'nunique','Spoiled Cost':'sum',\
                                               'Product Cost':'sum',\
                                               'Machine Inventory Cost': 'sum'}).sort_values('Total Collections',ascending=False)
```

```python
asset_df.groupby('Manufacturer',as_index=False).agg({'Spoiled Cost' : 'sum'}).sort_values('Spoiled Cost',ascending=False)
```
